```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('D:/Amazon Data Visualization/Sample_Longlist_Data.xlsx')
df
```

```python
plt.figure(figsize=(14, 8))
top_sellers = df[['seller business name', 'Hero Product 1 #ratings']].sort_values(by='Hero Product 1 #ratings', ascending=False).head(10)
sns.barplot(x='Hero Product 1 #ratings', y='seller business name', data=top_sellers, palette='viridis')
plt.title('Top 10 Sellers by Hero Product 1 #ratings')
plt.xlabel('Hero Product 1 #ratings')
plt.ylabel('Seller Business Name')
plt.show()
```

```python
# Distribution of Negative Ratings
plt.figure(figsize=(14, 8))
negative_ratings = df[['seller business name', 'Max % of negative seller ratings - last 30 days', 
                       'Max % of negative seller ratings - last 90 days', 
                       'Max % of negative seller ratings - last 12 months']].sort_values(by='Max % of negative seller ratings - last 12 months', ascending=False).head(10)

negative_ratings.set_index('seller business name').plot(kind='bar', stacked=True, figsize=(14, 8), cmap='viridis')
plt.title('Negative Seller Ratings Over Time')
plt.xlabel('Seller Business Name')
plt.ylabel('Percentage of Negative Ratings')
plt.xticks(rotation=45)
plt.legend(title='Time Period')
plt.show()
```

```python
# Seller Performance Overview (KPI Example)
kpi_data = df[['seller business name', 'Count of seller brands', 'sellerproductcount', 'sellerratings']].head(10)
print(kpi_data)
```

```python
# Product Category Analysis
plt.figure(figsize=(10, 6))
category_data = df['category'].value_counts()
category_data.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis', len(category_data)))
plt.title('Product Category Distribution')
plt.ylabel('')
plt.show()
```

```python
# Time-Based Analysis of Sellers
plt.figure(figsize=(14, 8))
time_data = df.groupby('Date Added').agg({'Hero Product 1 #ratings': 'sum'}).reset_index()
sns.lineplot(x='Date Added', y='Hero Product 1 #ratings', data=time_data)
plt.title('Growth of Seller Ratings Over Time')
plt.xlabel('Date Added')
plt.ylabel('Total Hero Product 1 #ratings')
plt.show()
```

```python
# Best Seller Identification
best_sellers = df[['seller business name', 'Hero Product 1 #ratings', 'Max % of negative seller ratings - last 12 months', 'Count of seller brands']]
best_sellers['Performance Score'] = best_sellers['Hero Product 1 #ratings'] - best_sellers['Max % of negative seller ratings -last 12 months'] * 10 + best_sellers['Count of seller brands'] * 5
best_sellers = best_sellers.sort_values(by='Performance Score', ascending=False).head(10)

plt.figure(figsize=(14, 8))
sns.barplot(x='Performance Score', y='seller business name', data=best_sellers, palette='viridis')
plt.title('Top 10 Sellers by Performance Score')
plt.xlabel('Performance Score')
plt.ylabel('Seller Business Name')
plt.show()
```

```python
import numpy as np

# Prepare data for Radar Chart
from math import pi

top_5_sellers = df[['seller business name', 'Count of seller brands', 'Hero Product 1 #ratings', 'Max % of negative seller ratings - last 12 months']].head(5)
top_5_sellers = top_5_sellers.set_index('seller business name')
categories = list(top_5_sellers.columns)
N = len(categories)

# Radar chart
values = top_5_sellers.iloc[0].tolist()
values += values[:1]
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories)

ax.plot(angles, values)
ax.fill(angles, values, 'b', alpha=0.1)

plt.title(f'Performance of {top_5_sellers.index[0]} Across Metrics')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


top_sellers = df.groupby('sellerlink').sum().sort_values(by=['Hero Product 1 #ratings'], ascending=False).head(5)
relevant_sellers = df[df['sellerlink'].isin(top_sellers.index)]

categories = ['Hero Product 1 Ratings', 'Hero Product 2 Ratings', 'Max % Negative Ratings (Last 12 Months)', 'Count of Seller Brands']
radar_data = []

for seller in relevant_sellers['sellerlink'].unique():
    seller_data = relevant_sellers[relevant_sellers['sellerlink'] == seller]
    values = [
        seller_data['Hero Product 1 #ratings'].values[0],
        seller_data['Hero Product 2 #ratings'].values[0],
        seller_data['Max % of negative seller ratings - last 12 months'].values[0],
        seller_data['Count of seller brands'].values[0]
    ]
    radar_data.append(values)

radar_data = np.array(radar_data)
radar_data = radar_data / radar_data.max(axis=0)


num_vars = len(categories)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()


angles += angles[:1]

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

axes[0].pie(relevant_sellers['Count of seller brands'], labels=relevant_sellers['sellerlink'], autopct='%1.1f%%', startangle=140)
axes[0].set_title('Distribution of Seller Brands')

# Plot 2: Radar Chart
ax2 = plt.subplot(122, polar=True)
for i, seller in enumerate(relevant_sellers['sellerlink'].unique()):
    values = radar_data[i].tolist()
    values += values[:1]
    ax2.plot(angles, values, label=seller)
    ax2.fill(angles, values, alpha=0.25)

ax2.set_title('Radar Chart Comparison of Sellers')
ax2.set_xticks(angles[:-1])
ax2.set_xticklabels(categories)
ax2.set_yticklabels([])  
ax2.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.tight_layout()
plt.show()
```
